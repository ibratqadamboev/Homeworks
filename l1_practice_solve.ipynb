```python
# Loading climate data from abu_dhabi_climate.csv file
df = pd.read_csv('./abu_dhabi_climate.csv')
df.head(10)
```

As we have seen from the dataset, it is the climate data of the city Abu Dhabi which is the one of the most prestigious cities of the United Arab Emirates. Because of this data, we can visualize some of the trends of the climate of the Arab country and find solution to some hyphothesises.

From my understanding of the climate, Arab countries are hottest countries in the world. Some statistics say that Arab countries are cooling because of climate change. Check if the rumors around this climate factor is correct or not.

```python
# Visualize the annual temperture change graph in linegraph
plt.plot(df['Year'], df['T'], color = "#ca0933")
plt.title("Annual Temperature change in UAE")
plt.xlabel("Years")
plt.ylabel("Annual temperature")
plt.show()
```

To get full understanding of temperature change, we have to also visualize graph of maximum temperature and minimum temperature change in different colors(with legend).

```python
# Visualize max, min temperature change in the same graph
plt.Figure(figsize = (8, 6))
plt.plot()
plt.title("Annual Temperature change in UAE")
plt.plot(df['TM'], color = "#12830fbe", label = "Maximum Temperature")
plt.plot(df['T'], color = "#cd2342", label = "Annual Temperature")
plt.plot(df['Tm'], color = "#c2cd23", label = "Minimum Temperature")
plt.legend(loc = 0)
plt.xlabel("Years")
plt.ylabel("Annual temperature")
```

There is also rumors in terms of increase in the rainy days of Arab peninsula. That is really interesting question as most people relate this factor to Doomsday(or Qiyamah(Islam)). Let's check if this rumor is correct or not with data we have.

```python
# Plotting overall precipation change line graph
plt.Figure(figsize = (8, 6))
plt.plot(df['Year'], df['PP'], color = "#d83535bc")
plt.title("Annual Temperature change in UAE")
plt.xlabel("Years")
plt.ylabel("Precipiation trends")
```

```python
# Plot rainy days(RA) and snowy days(SN) in different line
plt.figure(figsize = (8, 6))
plt.plot(df['Year'], df['RA'], color = "#207e13bb", label = "Rainy days")
plt.plot(df['Year'], df['PP'], color = "#d83535bc", label = "Overall")
plt.plot(df['Year'], df['SN'], color = "#d3d61cb9", label = "Snowy days")
plt.title("Annual Precipation change in UAE")
plt.xlabel("Years")
plt.ylabel("Precipation")
plt.legend()
plt.show()
```

From quick analysis of the dataset, we can see that there was storms, foggy days, tornado and other unusual climate spikes in the history of the climate of Arab peninsula. With the help of visulization of the dataset, try to answer following question?

* Which type of unusual climate spikes occured most often.

```python
# Visualize trends in Storms, Fog, Tornado and Hail in Arab peninsula
plt.figure(figsize = (8, 6))
plt.plot(df['Year'], df['TS'], color = "#207e13bb", label = "Storms")
plt.plot(df['Year'], df['FG'], color = "#c91a1aba", label = "Fog")
plt.plot(df['Year'], df['TN'], color = "#bec02fba" , label = "Tornado")
plt.plot(df['Year'], df['GR'], color = "#0abadaba" , label = "Hail")
plt.title("Annual Precipation change in UAE")
plt.xlabel("Years")
plt.ylabel("Number of occurence")
plt.legend(loc = 0)
plt.show()
```
