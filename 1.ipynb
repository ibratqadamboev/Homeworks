```python
import pygame
import random

pygame.init()

WIDTH, HEIGHT = 800, 600
CUBE_SIZE = 15
CUBE_WIDTH = WIDTH // CUBE_SIZE
CUBE_HEIGHT = HEIGHT // CUBE_SIZE
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Snake game")

clock = pygame.time.Clock()
SPEED = 15
snake = [(CUBE_WIDTH // 2, CUBE_HEIGHT // 2)]
snake_dir = (1,0)
food = (random.randint(0, CUBE_WIDTH - 1), random.randint(0, CUBE_HEIGHT - 1))
running = True
while running:
    for event in pygame.event.get() :
        if event.type == pygame.KEYDOWN :
            if event.key == pygame.K_w and snake_dir != (0, 1) :
                snake_dir = (0, -1)
            elif event.key == pygame.K_s and snake_dir != (0, -1) :
                snake_dir = (0, 1)
            elif event.key == pygame.K_a and snake_dir != (1, 0) :
                snake_dir = (-1, 0)
            elif event.key == pygame.K_d and snake_dir != (-1, 0) :
                snake_dir = (1,0)
    screen.fill((102, 255, 255)) 

    head = (snake[0][0] + snake_dir[0], snake[0][1] + snake_dir[1])

    if (head[0]  >= CUBE_WIDTH) or (head[0] <= 0) or (head[1] >= CUBE_HEIGHT) or (head[1] <= 0):
        running = False

    if head in snake :
        running = False

    snake.insert(0, head)

    if snake[0] == food :
        food = (random.randint(0, CUBE_WIDTH - 1), random.randint(0, CUBE_HEIGHT - 1))
    else :
        snake.pop()

        
    pygame.draw.rect(screen, (255, 0, 0), (food[0] * CUBE_SIZE, food[1] * CUBE_SIZE, CUBE_SIZE, CUBE_SIZE))
    for segment in snake :
        pygame.draw.rect(screen, (0, 255, 0), (segment[0] * CUBE_SIZE, segment[1] * CUBE_SIZE, CUBE_SIZE, CUBE_SIZE))


    pygame.display.flip()
    clock.tick(SPEED)

pygame.quit()
```
