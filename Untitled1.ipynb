```python
import csv
```

```python
with open("./safiya_raw_data.csv" ,"r" , encoding = "utf-8") as f:
  reader = csv.DictReader(f)
  row = list(reader)
row
```

```python
data = [ dict(item) for item in row]
data
```

```python
clean_data = [item for item in data if all(item.values())]
```

```python
uniquedata =[]
for item in clean_data:
  if item not in uniquedata:
    uniquedata.append(item)
```

```python
columns = uniquedata[0].keys()
columns
```

```python
for item in uniquedata :
    print(f"{item["customer_name"]}")
```

```python
branches = set(item["branch"] for item in uniquedata)
branches
```

```python
asosiy_branch = []
for item in uniquedata :
    if item["branch"] == "Asosiy" :
        asosiy_branch.append(item)
asosiy_branch
```

```python
sales_count = {}
for item in asosiy_branch :
    product = item["product_name"]
    amount = int(float(item["quantity"]))
    if product in sales_count :
        sales_count[product] += amount
    else :
        sales_count[product] = amount

sales_count
```

```python
Chilonzor_Branch = []
for item in uniquedata :
    if item["branch"] == "Chilonzor" :
        Chilonzor_Branch.append(item)
Chilonzor_Branch
```

```python
maximum = {}
for item in Chilonzor_Branch :
    product_name = item["product_name"]
    sotilish_soni = int(float(item["quantity"]))
    if product_name in maximum :
        maximum[product_name] += sotilish_soni
    else :
        maximum[product_name] = sotilish_soni

maximum
```

```python
max_product = max(maximum)
max_product
```
