```python
# Sinuoidal function graph
x = np.arange(0, 10, 0.1)
y = np.sin(x)

# Graphs using x, y lists
plt.plot(x, y)
plt.show()
```

**Different parameters**

linewidth – толщина линии (от 0.5 до 3)

linestyle – стиль линии

    '-' – сплошная линия

    '--' – пунктирная линия

    '-. ' – штрихпунктирная линия

    ':' – точечная линия

None или '' – без линии

color – цвет линии

    'b' – синий (blue)

    'c' – голубой (cyan)

    'g' – зелёный (green)

    'k' – чёрный (black)

    'm' – фиолетовый / пурпурный (magenta)

    'r' – красный (red)

    'y' – жёлтый (yellow)

Также можно задать цвет в формате HEX ('#FF5733') или RGBA ((255,182,193,0.5))

alpha – прозрачность линии (0 – полностью прозрачная, 0.5 – полупрозрачная, 1 – непрозрачная)

marker – маркер (отметка на точках)

    '.' – точка

    'o' – круг

    'v' – треугольник вниз

    '^' – треугольник вверх

    '*' – звезда

```python
# Graphs using x, y lists
plt.plot(x, y, linewidth =  1, linestyle = ":")
plt.show()
```

```python
# Graphs using x, y lists
plt.plot(x, y, linewidth = 4, linestyle = ":", color = "#1BFF8D", alpha = 0.3, marker = 'v')
plt.show()
```

```python
# Graphs using x, y lists
plt.title("Sin(x) funksiya")
plt.plot(x, y, linewidth = 4, linestyle = ":", color = '#8fff33', alpha = 0.3, marker = 'v')
plt.ylabel("Sin(x)")
plt.xlabel("x")
plt.grid()
plt.show()
```

```python
# Multiple graphs of the functions
plt.figure(figsize = (12, 4))
plt.title("Sin(x) funksiya")
plt.plot(x, y, linewidth = 4, linestyle = ":", color = '#8fff33', alpha = 0.3, marker = 'v')
plt.plot(x + 1, y, linewidth = 2, linestyle = "-.", color = '#5e33ff', alpha = 0.9, marker = 'o')
plt.ylabel("Sin(x)")
plt.xlabel("x")
plt.grid()
plt.show()
```

```python
# Multiple graphs of the functions
plt.figure(figsize = (12, 4))
plt.title("Sin(x) funksiya")
plt.plot(x, y, linewidth = 4, linestyle = ":", color = '#8fff33', alpha = 0.3, marker = 'v', label = 'sin(x)')
plt.plot(x + 1, y, linewidth = 2, linestyle = "-.", color = '#5e33ff', alpha = 0.9, marker = 'o', label = "sin(x+1)")
plt.legend(loc = 0, fontsize='large')
plt.ylabel("Sin(x)")
plt.xlabel("x")
plt.grid()
plt.savefig("practice.png")
```
