```python
# Pie chart implementation
sex_of_people = df['sex'].value_counts()

plt.figure(figsize = (12, 6))
plt.title("Number of people based on their sex")
plt.pie(x = sex_of_people.values, labels = ['Male', 'Female', 'Other'])
plt.show()
```

```python
# Pie chart implementation
smoke = df['smoke'].value_counts()

plt.figure(figsize = (12, 6))
plt.title("Number of smokers and non-smokers")
plt.pie(x = smoke.values, labels = ["Don't smoke", 'Smoke'], colors = ["#02f64f", "#ff0000"], shadow = True, rotatelabels = True, radius = 0.5, autopct='%1.1f%%')
plt.show()
```

**Correlation and Heatmaps**

Correlation is a statistical measure that describes the strength and direction of a relationship between two or more variables in a dataset. It indicates whether an increase or decrease in one variable is associated with a corresponding change in another variable

* A positive correlation means that as one variable increases, the other tends to increase as well.

* A negative correlation means that as one variable increases, the other tends to decrease.

* A correlation near zero suggests little to no linear relationship between the variables.

```python
# Checking Pearson correlation between variables
df.corr()
```

```python
# Drawing correlation matrix using heatmaps
plt.figure(figsize = (12, 6))
plt.title("Heatmap for correlation matrix", pad = 20)
plt.matshow(df.corr(), cmap = 'coolwarm', fignum = 1)

plt.xticks(ticks=np.arange(len(df.corr().columns)), labels=df.corr().columns, rotation = 90)
plt.yticks(ticks=np.arange(len(df.corr().columns)), labels=df.corr().columns)

plt.colorbar()

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

# Drawing correlation matrix using heatmaps
plt.figure(figsize = (12, 6))
plt.title("Heatmap for correlation matrix")

sns.heatmap(df.corr(), cmap = 'coolwarm', annot = True, fmt=".2f", linewidths=0.5)
plt.show()
```
